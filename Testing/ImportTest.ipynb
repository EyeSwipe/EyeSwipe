```python
import tensorflow as tf
import cv2
import numpy as np
import os
```

```python
VIDEOPATH = '/home/max/Desktop/TensorFlow/EyeSwipe/Data/I_know.MOV'
```

```python
# Check that the video exists
if not os.path.isfile(VIDEOPATH):
    print('Failed to find file at path: ' + VIDEOPATH)
```

```python
# adapted from:
# https://towardsdatascience.com/deep-learning-with-tensorflow-part-4-face-classification-and-video-inputs-fa078f22c1e5
# and:
# https://stackoverflow.com/a/4282857
def collect_video(video_path):
    video_capture = cv2.VideoCapture(video_path)
    
    imgs = []
    
    while True:
        frame = video_capture.read()[1] # get next frame
        if frame is None:
            break

        np_image_data = np.asarray(frame)

        imgs.append(np_image_data)
    
    return tf.convert_to_tensor(imgs)
```

```python
collect_video(VIDEOPATH)
```
